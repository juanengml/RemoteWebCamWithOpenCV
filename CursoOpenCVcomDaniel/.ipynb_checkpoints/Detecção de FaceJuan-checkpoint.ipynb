```python
classificadorFace = cv2.CascadeClassifier("cascades/haarcascade_frontalface_default.xml")
```

```python
classificadorOlhos = cv2.CascadeClassifier("cascades/haarcascade_eye.xml")
```

```python
def showImage(imagem):
 plt.imshow(imagem)
 plt.title('my picture')
 plt.show()
```

```python
imagem = cv2.imread("pessoas/pessoas2.jpg")
imagemCinza = cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
```

```python
facesDetectadas = classificadorFace.detectMultiScale(imagemCinza)
```

```python


for (x ,y , l ,a ) in facesDetectadas:
    imagem = cv2.rectangle(imagem, (x,y), (x +l, y + a),(100,100,255),2)
    regiao = imagem[y:y + a, x:x + l]
    regiaoCinzaOlho = cv2.cvtColor(regiao, cv2.COLOR_BGR2GRAY)
    olhosDetectados = classificadorOlhos.detectMultiScale(regiaoCinzaOlho,scaleFactor=1.11,minNeighbors=2)
    print (olhosDetectados)
    for  (ox , oy, ol, oa) in olhosDetectados:
          cv2.rectangle(regiao, (ox, oy), (ox + ol, oy + oa), (255,0,255),2)
```

```python
showImage(imagem)
```

## WebCam  Stream

```python
import cv2
```

```python
video = cv2.VideoCapture(0)
```

```python
while True:
    connectada, frame = video.read()
    cv2.imshow("Face",frame)
    if cv2.waitKey(1)  == ord('q'):
        break;


video.release()
cv2.destroyAllWindows()        
```
